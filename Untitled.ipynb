### Soal 1
Diketahui tinggi badan siswa kelas A adalah 100,5 ; 99,7 ; 101 ; 101,3 ; 99,9 ; 98,9 ; 165 ; 102,2 ; 98,7 ; 102,9 ; dan 100. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

##### mean

```python
data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

jumlah = sum(data)
print(f"Jumlah semua nilai: {jumlah}")

total_data = len(data)
print(f"Banyaknya data ada : {total_data}")

mean = jumlah / total_data
print(f"Rata-Ratanya adalah : {mean}")

import matplotlib.pyplot as plt

plt.hist(data, bins=35, edgecolor='blue', alpha=1, color='blue')

plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean:.2f}')

plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.title('Mean')

plt.legend()

plt.show()
```

#### median

```python
data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

sort = sorted(data)
print(f"Data berurut: {sort}")

x = len(sort)
print(f"Total data: {x}")

index_tengah = x//2 

if x % 2 == 0:
    print(f"Jumlah data Genap")
    med1 = sort[index_tengah]
    med2 = sort[index_tengah - 1]
    print(f"Nilai tengah: {med1} {med2}")
    median_value = (med1 + med2) / 2
else:
    print(f"Jumlah data Ganjil")
    median = sort[index_tengah]
    print(f"Nilai tengah: {median}")

print(f"Median: {median}")
```

```python
plt.hist(data, bins=35, edgecolor='black', alpha=0.5, color='blue')

plt.axvline(x=median, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median:.2f}')

plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.title('Median')

plt.legend()

plt.show()
```

#### modus

```python
from collections import Counter

data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

counter = Counter(data)
mode_data = dict(counter) 
max_counter = max(list(counter.values()))
print(f"Paling banyak muncul ada : {max_counter} kali")

mode_value1 = []
for k, v in mode_data.items():
    print(f"{k} muncul {v} kali")
    if v == max(list(counter.values())):
        mode_value1.append(k)
    
print(f"Mode: {mode_value1}")

```

```python
import matplotlib.pyplot as plt

plt.hist(data, bins=30, edgecolor='black', alpha=1, color='blue')

for k in mode_value1:
    plt.axvline(x=k, color='red', linestyle='dashed', linewidth=1, label=f'Mode = {k:.2f}')

plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.title('Modus')

plt.legend()

plt.show()
```

### Soal 2
Berikut adalah nilai UAS pada mata pelajaran matematika di kelas A: 95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

#### mean

```python
data = [95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75]
print(f"Data: {data}")

jumlah = sum(data)
print(f"Jumlah semua nilai: {jumlah}")

total_data = len(data)
print(f"Banyaknya data ada : {total_data}")

mean = jumlah / total_data
print(f"Rata-Ratanya adalah : {mean}")

import matplotlib.pyplot as plt

plt.hist(data, bins=35, edgecolor='blue', alpha=0.5, color='blue')

plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean:.2f}')

plt.xlabel('Nilai Siswa')
plt.ylabel('Frekuensi')
plt.title('Mean')

plt.legend()

plt.show()
```

#### median

```python
data = [95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75]
print(f"Data: {data}")

sort = sorted(data)
print(f"Data berurut: {sort}")

x = len(sort)
print(f"Total data: {x}")

index_tengah = x//2 

if x % 2 == 0:
    print(f"Jumlah data Genap")
    med1 = sort[index_tengah]
    med2 = sort[index_tengah - 1]
    print(f"Nilai tengah: {med1} {med2}")
    median_value = (med1 + med2) / 2
else:
    print(f"Jumlah data Ganjil")
    median = sort[index_tengah]
    print(f"Nilai tengah: {median}")

print(f"Median: {median}")

plt.hist(data, bins=35, edgecolor='black', alpha=0.5, color='blue')

plt.axvline(x=median_value, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median_value:.2f}')

plt.xlabel('Nilai Siswa')
plt.ylabel('Frekuensi')
plt.title('Median')

plt.legend()

plt.show()
```

#### modus
